```python
import matplotlib.pyplot as plt

temperature = [22, 27, 28, 29, 26]
days = ["sun", "mon", "tues", "wed", "thur"]

plt.barh(days, temperature, color='blue')  # hatch adds density-like pattern

plt.title("Maximum temperature in a week")
plt.xlabel("degree celsius")
plt.ylabel("day")

plt.show()
```

```python
import matplotlib.pyplot as plt

temperature = [22, 27, 28, 29, 26]
days = ["sun", "mon", "tues", "wed", "thur"]

plt.scatter(days, temperature, color='blue')  # hatch adds density-like pattern

plt.title("Maximum temperature in a week")
plt.xlabel("degree celsius")
plt.ylabel("day")

plt.show()
```
